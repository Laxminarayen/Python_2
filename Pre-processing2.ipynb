```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dataf =  pd.read_csv("dataf.csv")
```

```python
import pandas_profiling
```

```python
dataf = dataf.drop(columns = ["BsmtFinSF2","BsmtHalfBath","PoolArea","ScreenPorch","MiscVal","LowQualFinSF","3SsnPorch","EnclosedPorch"],axis = 0)
```

```python
#from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
dataf.TotalBsmtSF = pd.to_numeric(dataf.TotalBsmtSF,errors = 'raise')
```

```python
dataf.GarageType.unique()
```

```python
def encode(x):
    if x ==0:
        return 1
    elif x <1000:
        return 2
    else:
        return 3
```

```python
dataf["2ndFlrSF"] = dataf["2ndFlrSF"].transform(encode)
```

```python
dataf["2ndFlrSF"]
```

```python
dataf["BsmtFinSF1"].unique()
```

```python
dataf["BsmtFinSF1"] = dataf["BsmtFinSF1"].transform(encode)
```

```python
dataf["BsmtFinSF1"]
```

```python
def encode1(x):
    if x == 0:
        return "One"
    elif x == 1:
        return "Two"
    else:
        return "MORE than one"
```

```python
dataf["BsmtFullBath"] = dataf["BsmtFullBath"].transform(encode1)
```

```python
dataf["BsmtFullBath"]
```

```python
dataf["HalfBath"] = dataf["HalfBath"].transform(encode1)
```

```python
def encode2(x):
    if x == 0:
        return "0"
    if x == 1:
        return "1"
    elif x == 2:
        return "2"
    else:
        return ">2"
```

```python
dataf["Fireplaces"] = dataf["Fireplaces"].transform(encode2)
```

```python
dataf["Fireplaces"]
```

```python
dataf["WoodDeckSF"] = dataf["WoodDeckSF"].transform(encode)
```

```python
dataf["MasVnrArea"] = dataf["MasVnrArea"].transform(encode)
```

```python
dataf["MasVnrArea"]
```

```python
def encode4(x):
    if x ==0:
        return 1
    elif x < 500:
        return 2
    else:
        return 3
```

```python
dataf["OpenPorchSF"] = dataf["OpenPorchSF"].transform(encode4)
```

```python
data_column_category = dataf.select_dtypes(exclude=[np.number]).columns
```

```python
data_column_Integer = dataf.columns.difference(data_column_category)
```

```python
dataf_one_hot = pd.get_dummies(dataf[data_column_category])
```

```python
dataf_one_hot
```

```python
data_model = pd.concat([dataf_one_hot,dataf[data_column_Integer]],axis=1)
```

```python
data_model
```

```python
data_model.columns
```

```python
train_modified = data_model.loc[data_model.source_train==1.0,:]
```

```python
train_modified.shape
```

```python
test_modified = data_model.loc[data_model.source_train==0.0,:]
```

```python
test_modified.shape
```

```python
test_modified.drop(columns=['source_test', 'source_train'],inplace=True)
```

```python
train_modified.drop(columns=['source_test', 'source_train'],inplace=True)
```

```python
train_modified.to_csv("train_ready_for_model_1.csv",index=False)
```

```python
test_modified.to_csv("test_ready_for_model_1.csv",index=False)
```

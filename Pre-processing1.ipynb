```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train["source"] = "train"
test["source"] = "test"
```

```python
dataf = pd.concat([train,test],axis = 0,sort = False)
```

```python
dataf.info()
```

```python
dataf.LotFrontage.unique()
```

```python
def applymean(x):
    a = x.mean()
    print(a)
    return x.fillna(int(a))
```

```python
def applymode(x):
    a = x.mode()[0]
    print(a)
    return x.fillna(a)
```

```python
dataf["LotFrontage"] = dataf.LotFrontage.transform(applymean)
```

```python
dataf.Utilities.unique()
```

```python
dataf["Utilities"] = dataf.groupby("MSSubClass").Utilities.transform(applymode)
```

```python
#imputer loop
for i in data_column_category:
    dataf.loc[:,i] = dataf.loc[:,i].transform(applymode)
```
